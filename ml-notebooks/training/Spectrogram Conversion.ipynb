```python
import librosa
import librosa.display
from zipfile import ZipFile
from scipy.io import wavfile as wav
import numpy as np
import pandas as pd
from scipy.io import wavfile as wav
import matplotlib.pyplot as plt
import os
```

```python
folder = "D:\\Dataset\\BabyCry_Clean"
for count, filename in enumerate(os.listdir(folder)):
    audio_file_path = folder + "\\" + filename
    print(audio_file_path)
    audioData, sampleRate = librosa.load(audio_file_path)
    waveSampleRate, audio = wav.read(audio_file_path)
    spec = librosa.stft(audioData)
    spec_mag, _ = librosa.magphase(spec)
    melScaleSpec = librosa.feature.melspectrogram(S=spec_mag, sr=waveSampleRate)
    melSpec = librosa.amplitude_to_db(melScaleSpec, ref=np.min)
    librosa.display.specshow(melSpec, sr=sampleRate, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')  
    plt.title('BabyCryClean')
    plt.savefig("D:\\Spectrogram Dataset\\BabyCry_Clean\\BabyCryClean_" + str(count) + ".png")
    plt.clf()
```
